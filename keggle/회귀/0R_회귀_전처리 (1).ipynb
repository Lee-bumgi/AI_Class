```python
# ADDRESS 드랍
train = train.drop(['ADDRESS'], axis=1)
test = test.drop(['ADDRESS'], axis=1)
# SUBURB 드랍
train = train.drop(['SUBURB'], axis=1)
test = test.drop(['SUBURB'], axis=1)

# GARAGE 결측치 처리
train['GARAGE'] = train['GARAGE'].fillna(0)
test['GARAGE'] = test['GARAGE'].fillna(0)
train['GARAGE'].replace(99,2,inplace=True) # test에는 99없음
# 제곱 컬럼 추가
train['GARAGE_2'] = train.GARAGE**2
test['GARAGE_2'] = test.GARAGE**2

# LAND_AREA 이상치 처리
train['LAND_AREA'].replace(999999,692,inplace=True)
train['LAND_AREA'].replace(983690,692,inplace=True)
test['LAND_AREA'].replace(999999,692,inplace=True)
test['LAND_AREA'].replace(983690,692,inplace=True)
# train['LAND_AREA'] = np.log(train['LAND_AREA'] + 1)
# test['LAND_AREA'] = np.log(test['LAND_AREA'] + 1)
# 로그컬럼추가
train['LAND_AREA_log'] = np.log(train.LAND_AREA+1)
test['LAND_AREA_log'] = np.log(test.LAND_AREA+1)

# FLOOR_AREA 제곱컬럼 추가
train['FLOOR_AREA_2'] = train.FLOOR_AREA ** 2
test['FLOOR_AREA_2'] = test.FLOOR_AREA ** 2 

# BUILD_YEAR 결측치 랜덤 넣기
train['BUILD_YEAR_random'] = train['BUILD_YEAR']
test['BUILD_YEAR_random'] = test['BUILD_YEAR']
temp_train = (train['BUILD_YEAR'].dropna().sample(train['BUILD_YEAR'].isnull().sum()))
temp_test = (test['BUILD_YEAR'].dropna().sample(test['BUILD_YEAR'].isnull().sum()))
temp_train.index = train[lambda x: x['BUILD_YEAR'].isnull()].index # index 부여
temp_test.index = test[lambda x: x['BUILD_YEAR'].isnull()].index # index 부여
train.loc[train['BUILD_YEAR'].isnull(), 'BUILD_YEAR_random'] = temp_train
test.loc[test['BUILD_YEAR'].isnull(), 'BUILD_YEAR_random'] = temp_test
train[['BUILD_YEAR', 'BUILD_YEAR_random']].isnull().sum()
test[['BUILD_YEAR', 'BUILD_YEAR_random']].isnull().sum()
# BUILD_YEAR 드랍
train = train.drop(['BUILD_YEAR'], axis=1)
test = test.drop(['BUILD_YEAR'], axis=1)


# CBD_DIST 제곱추가
train['CBD_DIST_2'] = train.CBD_DIST**2
test['CBD_DIST_2'] = test.CBD_DIST**2

# NEAREST_STN 드랍
train = train.drop(['NEAREST_STN'], axis=1)
test = test.drop(['NEAREST_STN'], axis=1)

# NEAREST_STN_DIST 제곱추가
train['NEAREST_STN_DIST_2'] = train.NEAREST_STN_DIST**2
test['NEAREST_STN_DIST_2'] = test.NEAREST_STN_DIST**2
# NEAREST_STN_DIST 드랍
# train = train.drop(['NEAREST_STN_DIST'], axis=1)
# test = test.drop(['NEAREST_STN_DIST'], axis=1)

# DATE_SOLD 문자열 -> 정수형
def split_m(row):
    for i in range(train['DATE_SOLD'].size):
        row = row[i][:7]
        month, year = row.split('-')
        return int(year+month)
train['DATE_SOLD_cut'] = train[['DATE_SOLD']].apply(split_m,axis=1)
test['DATE_SOLD_cut'] = test[['DATE_SOLD']].apply(split_m,axis=1)
# DATE_SOLD 년도, 월 분리 하여 새컬럼 만들기
def split_year(row):
    for i in range(train['DATE_SOLD'].size):
        row = row[i][:7]
        month, year = row.split('-')
        return int(year)
def split_month(row):
    for i in range(train['DATE_SOLD'].size):
        row = row[i][:7]
        month, year = row.split('-')
        return int(month)
train['DATE_YEAR'] = train[['DATE_SOLD']].apply(split_year,axis=1)
# train['DATE_MONTH'] = train[['DATE_SOLD']].apply(split_month,axis=1)
test['DATE_YEAR'] = test[['DATE_SOLD']].apply(split_year,axis=1)
# test['DATE_MONTH'] = test[['DATE_SOLD']].apply(split_month,axis=1)
train = train.drop(['DATE_SOLD'], axis=1)
test = test.drop(['DATE_SOLD'], axis=1)

# POSTCODE 드랍
train = train.drop(['POSTCODE'], axis=1)
test = test.drop(['POSTCODE'], axis=1)

# NEAREST_SCH_DIST
train['NEAREST_SCH_DIST_2'] = train.NEAREST_SCH_DIST**2
test['NEAREST_SCH_DIST_2'] = test.NEAREST_SCH_DIST**2
# train = train.drop(['NEAREST_SCH_DIST'], axis=1)
# test = test.drop(['NEAREST_SCH_DIST'], axis=1)

# NEAREST_SCH 대학 단과대학 학교 기타 분류
def school_split(row):
    for i in range(train['NEAREST_SCH'].size):
        if ('SCHOOL' in row[i]) and ('COMMUNITY' in row[i]):
            return 'communityschool'
        elif ('SCHOOL' in row[i]) and ('COMMUNITY' not in row[i]) :
            return 'school'
        elif ('COLLEGE' in row[i]) and ('SENIOR' in row[i]):
            return 'senior college'
        elif ('COLLEGE' in row[i]) and ('SENIOR' not in row[i]):
            return 'college'
        else:
            return 'other'
# NEAREST_SCH 드랍
# train = train.drop(['NEAREST_SCH'], axis=1)
# test = test.drop(['NEAREST_SCH'], axis=1)
train['NEAREST_SCH'] = train[['NEAREST_SCH']].apply(school_split,axis=1)
test['NEAREST_SCH'] = test[['NEAREST_SCH']].apply(school_split,axis=1)

# NEAREST_SCH_RANK 드랍
train['NEAREST_SCH_RANK_random'] = train['NEAREST_SCH_RANK']
test['NEAREST_SCH_RANK_random'] = test['NEAREST_SCH_RANK']
temp_train = (train['NEAREST_SCH_RANK'].dropna().sample(train['NEAREST_SCH_RANK'].isnull().sum()))
temp_test = (test['NEAREST_SCH_RANK'].dropna().sample(test['NEAREST_SCH_RANK'].isnull().sum()))
temp_train.index = train[lambda x: x['NEAREST_SCH_RANK'].isnull()].index # index 부여
temp_test.index = test[lambda x: x['NEAREST_SCH_RANK'].isnull()].index # index 부여
train.loc[train['NEAREST_SCH_RANK'].isnull(), 'NEAREST_SCH_RANK_random'] = temp_train
test.loc[test['NEAREST_SCH_RANK'].isnull(), 'NEAREST_SCH_RANK_random'] = temp_test
train[['NEAREST_SCH_RANK', 'NEAREST_SCH_RANK_random']].isnull().sum()
test[['NEAREST_SCH_RANK', 'NEAREST_SCH_RANK_random']].isnull().sum()
train = train.drop(['NEAREST_SCH_RANK'], axis=1)
test = test.drop(['NEAREST_SCH_RANK'], axis=1)
```

```python
gb = GradientBoostingRegressor(n_estimators=3000, learning_rate=0.05,
                                   max_depth=4, max_features='sqrt',
                                   min_samples_leaf=15, min_samples_split=10, 
                                   loss='huber')
```

```python
xgboost = xgb.XGBRegressor(colsample_bytree=0.4603, gamma=0.0468, 
                             learning_rate=0.05, max_depth=3, 
                             min_child_weight=1.7817, n_estimators=2200,
                             reg_alpha=0.4640, reg_lambda=0.8571,
                             subsample=0.5213,
                             random_state =7, nthread = -1)
```

```python
lightgb = lgb.LGBMRegressor(objective='regression',num_leaves=5,
                              learning_rate=0.05, n_estimators=7200,
                              max_bin = 55, feature_fraction_seed=9, bagging_seed=9,
                              n_jobs=-1)
```

```python
svr = make_pipeline(RobustScaler(), SVR(C= 20, epsilon= 0.008, gamma=0.0003))
```

```python
alpha = 0.0033000000000000004
lasso = LassoCV(alphas = [alpha * .6, alpha * .65, alpha * .7, alpha * .75, alpha * .8, 
                          alpha * .85, alpha * .9, alpha * .95, alpha, alpha * 1.05, 
                          alpha * 1.1, alpha * 1.15, alpha * 1.25, alpha * 1.3, alpha * 1.35, 
                          alpha * 1.4], 
                max_iter = 50000, cv = 10, n_jobs=-1)
lasso = make_pipeline(RobustScaler(),lasso)
```

```python
alpha = 0.6
ridge = RidgeCV(alphas = [alpha * .6, alpha * .65, alpha * .7, alpha * .75, alpha * .8, alpha * .85, 
                          alpha * .9, alpha * .95, alpha, alpha * 1.05, alpha * 1.1, alpha * 1.15,
                          alpha * 1.25, alpha * 1.3, alpha * 1.35, alpha * 1.4], cv = 10)
ridge = make_pipeline(RobustScaler(),ridge)
```
