```python
# %s 문자열
# %c 문자 한개
# %d 0~9까지의 정수
```

```python
day = 7
s = "오늘은 7월 %d일 입니다."%day
print(s)
```

```python
month = 7
day = 7
s = "오늘은 %d월 %d일 입니다."%(month,day)
print(s)
```

```python
d = "오늘은 {}월 {}일 입니다.".format(month,day)
d
```

## 삼항 연산자

```python
# 삼항 연산자
# a if 조건식 else b

score = 80
"합격" if score >= 60 else "불합격"
```

```python
a= int(input())
b = int(input())
print(f"a: {a}" if a>b else f"b: {b}")
```

```python
a= int(input())
b= int(input())
print(f"두 수의 차 : {a-b}" if a>b else f"두 수의 차 :{b-a}")
```
