# while문

```python
# while문은 반복횟수가 명확하지 않을때
# for문은 반복 횟수가 명확할 때
```

```python
while 조건식:
    실행문장
# 조건식에는 True or False가 들어갈 수 있도록 한다.
```

```python
number = 1
while number <= 3:
    print(number)
    number+=1
```

```python
a=1
while a<=10:
    print("파이썬 최고!!")
    a+=1
```

### break - break를 만나는 순간 그 자리에서 바로 반복문 바깥으로 빠져나간다
###              그래서 보통 조건문과 함께 사용된다.

```python
while True:
    print("무한루프")
    break
```

```python
number = 1
while True:
    print(number)
    number += 1
    if number > 3:
        break
```

### continue  -  밑에 코드는 실행되지 않고 반복문 위쪽으로 다시 돌아간다

```python
num = 1
while num <= 10:
    print(f"{num}번째 반복중..")
    if num > 4:
        print("여기서 부터는 num이 증가하지 않습니당..")
        continue
    num+=1
```

```python
while True:
    a = input("이름 입력 >> ")
    if a == "최성우":
        print(f"{a} 멋지다!")
        break
    print(f"{a} 최고!")
```

```python
while True:
    a = int(input('첫 번째 정수 입력 >> '))
    b = int(input('두 번째 정수 입력 >> '))
    if a ==0 and b ==0:
        print('프로그램이 종료되었습니다')
        break
    print(f"두 정수의 합 : {a+b}")
```

```python
current_weight = int(input("현재 몸무게 : "))
target_weight = int(input("목표 몸무게 : "))
n=1
while True:
    a = int(input("{}주차 감량 몸무게 : ".format(n)))
    current_weight -= a
    n+=1
    if current_weight <= target_weight:
        print("{}kg 달성!! 축하합니다!".format(current_weight))
        break    
```

```python
import random
a= random.randint(1,50)

while True:
    b= int(input("숫자를 입력하세요 >>"))
    if b < a:
        print(f"{b}보다 큰 수입니다.")
    elif b>a:
        print(f"{b}보다 작은 수입니다.")
    else:
        print("정답입니다~")
        break
```

# for문

```python
for 변수 in 문자열(or 리스트 or 튜플):
    print(변수)
```

```python
list_food = ['햄버거','치킨','피자']
for food in list_food:
    print(food)
```

```python
hi = '안녕하세요'
for s in hi:
    print(s)
```

```python
tuple_food = ('햄버거','치킨','피자')
for food in tuple_food:
    print(food)
```

```python
score_list = [90,45,70,60,55]
for i in score_list:
    if i >=60:
        print(f"{score_list.index(i)+1}번 학생은 합격입니다!")
    else:
        print(f"{score_list.index(i)+1}번 학생은 불합격입니다!")
```

```python
python_score = [57,86,63,92,35,79]
for i in python_score:
    if i >=80:
        print(f"{python_score.index(i)+1}번은 {i}점 이며, A등급 입니다.")
    elif i >=60:
        print(f"{python_score.index(i)+1}번은 {i}점 이며, B등급 입니다.")
    else:
        print(f"{python_score.index(i)+1}번은 {i}점 이며, C등급 입니다.")        
```

```python
python_score = [57,86,63,92,35,79]
num = 0

for i in python_score:
    if i >=80:
        grade = "A"
    elif i >=60:
        grade = "B"
    else:
        grade = "C"
    num+=1
    print(f"{num}번은 {i}점이며, {grade}등급입니다.")
```

```python
list1 = [1,30,7,9,17,3,5,23,44]
a=0
b=0
for i in range(len(list1)):
    if list1[i]%2==0:
        a+=1
    if list1[i]%2==1:
        b+=1
print("짝수 {}개, 홀수 {}개".format(a,b))
```

```python
inputOX = input("OX입력 >> ")
my_list = list(inputOX)
score = 0
extra = 0

for i in range(1,len(my_list)):
    if my_list[i]=='O':
        if my_list[i-1]=='O':
            extra +=1
        score +=1
        score +=extra
    else:
        extra = 0
if my_list[0] == 'O':
    score +=1
print(f"{score}점")
```

```python
# range(len(my_list))에서 list길이가 10일때 len(my_list)도 10이 되고, 이게 range에 들어가게 되면 9까지 되는데
# 그 9라는 수치는 인덱스 값으로 인식이 되기 때문에 끝까지인 10개 모두 출력이 된다.
```
