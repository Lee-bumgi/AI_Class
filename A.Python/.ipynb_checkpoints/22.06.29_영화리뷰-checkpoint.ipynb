# 네이버 영화리뷰 분석
### pickle 라이브러리

```python
# pickle : 리스트와 같은 자료구조 자체를 저장하거나 불러오게 해주는 라이브러리(기능들의 집합)
```

```python
import pickle
with open('Data/movie_review.pkl', 'rb') as f:    # rb: 읽기전용
    my = pickle.load(f)
my
```

```python
# 0 : 부정리뷰, 1 : 긍정리뷰
```

```python
print(len(my))
```

```python
for i in range(len(my)):
    print(my[i][0])

# for i in my:
#    print(i[0])
```

```python
a=[]
for i in my:
    a.append(len(i[0]))
b=max(a)
print(max(a))

for i in my:
    if len(i[0]) ==b:
        print(i[0])
```

```python
# 리뷰 최대값 / 내용 출력
max_len = len(my[0][0]) # 첫 번째 리뷰 길이
max_content = ""        # 가장 길이가 긴 리뷰가 들어갈 변수

for i in my:
    review_len = len(i[0])
    if max_len < review_len:
        max_len = review_len
        max_content = i[0]
print(f"리뷰 최대길이는 '{max_len}'입니다.")
print(max_content)
```

```python
length=''
for i,j in my:
    if len(i)>len(length):
        length=i
print(f"리뷰 최대길이는 '{len(length)}'입니다")
print(length)
```

```python
# i,j,k 도 되는가?
# 하나가 i만 있는 값이 있으면 어떻게 되는가?
```

```python
# 리뷰 길이 최소값/ 리뷰내용
min_len = len(my[0][0])
min_content = ""

for i in my:
    review_len = len(i[0])
    if min_len > review_len:
        min_len = review_len
        min_content = i[0]
print(f"리뷰 최소길이는 '{min_len}'입니다.")
print(min_content)
```

```python
length=my[0][0]
for i,j in my:
    if len(i)<len(length):
        length=i
print(f"리뷰 최소길이는 '{len(length)}'입니다")
print(length)
```

```python
# 리뷰 길이 평균값
a=[]
for i in my:
    a.append(len(i[0]))
b= sum(a)/len(a)
print(f"리뷰의 평균길이는 '{b}'입니다")
```

```python
total = 0
for i in my:
    total += len(i[0])
print(f"리뷰의 평균길이는 '{total/len(my)}'입니다.")
```

```python
# 긍정과 부정리뷰의 개수
pos_com = 0
neg_com = 0

for i in my:
    if i[1]==0:
        neg_com +=1
    else:
        pos_com +=1
print("긍정 '{}'개".format(pos_com))
print("부정 '{}'개".format(neg_com))
```

```python
# 두 번째 방법
re_list =[]    # 0또는 1 값 저장할 자리
for i in my:
    re_list.append(i[1])
neg = re_list.count(0)
pos = re_list.count(1)
print("긍정 '{}'개".format(pos))
print("부정 '{}'개".format(neg))
```
