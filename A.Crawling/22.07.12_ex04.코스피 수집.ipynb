```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
res = req.get("https://finance.naver.com/sise/sise_quant.naver")
soup = bs(res.text,'lxml')
tltle = soup.select('a.tltle')
cur_price_data = soup.select('td.number:nth-child(3)')
trade_data = soup.select('td.number:nth-child(6)')
title=[]
cur_price=[]
trade=[]
for i in tltle:
    title.append(i.text)
for i in cur_price_data:
    cur_price.append(i.text)
for i in trade_data:
    trade.append(i.text)
    
col =['종목명']
result = pd.DataFrame(title,columns=col)
result['현재가격']=cur_price
cur_price = pd.DataFrame(cur_price)
result['거래량']=trade
trade = pd.DataFrame(trade)
result
```

```python
result.to_csv("주식.csv",encoding = 'euc-kr')
```
