```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
# 1. 네이버 금융사이트에서 코스피 Top100사이트의 정보를 요청
res = req.get("https://finance.naver.com/sise/sise_quant.naver")
```

```python
res.text
```

```python
#2. 사용자가 받아온 데이터(text) 컴퓨터가 알아 먹을수 있게 컴퓨터용 언어로 변형!
# html언어로 변형 > bs라이브러리!
soup = bs(res.text,"lxml")
```

```python
#3. soup데이터에서 내가 필요한 요소를 찾아오기!
title = soup.select("a.tltle")
```

```python
price = soup.select("td.number:nth-child(3)")
```

```python
volume = soup.select("td.number:nth-child(6)")
```

```python
#4. 데이터의 개수가 정확하진 꼭 확인!
# 리스트의 데이터 길이를 확인! len()
len(price)
```

```python
# title,volume,price 리스트에서 순수한 글자정보만 추출한 리스트를 새로 제작
titleList = []
priceList = []
volumeList = []

for i in range(len(title)) :
    titleList.append(title[i].text)
    priceList.append(price[i].text)
    volumeList.append(volume[i].text)
```

```python
len(volumeList)
```

# 정제된(수순한 글자)데이터를 표 형식 제작

```python
import pandas as pd
```

```python
data = {"종목명" : titleList, "가격정보" : priceList, "거래량" : volumeList}
```

```python
kospi = pd.DataFrame(data)
```

```python
kospi
```

```python
# 1. euc-kr
# 2. ""
# 3. utf-8-sig
kospi.to_csv("코스피데이터.csv", encoding = "euc-kr")
```
