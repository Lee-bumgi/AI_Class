### 노래제목 수집

```python
# 노래제목에 해당하는 태그정보를 알아온 다음에,
# soup.select() 노래제목만 수집
```

```python
# 클래스 사이에 공백이 있다는 의미는 클래스가 복수개
# 공백을 지우고 , 를 통해서 클래스를 이어준다.
title = soup.select("div.ellipsis.rank01>span>a")
title
```

```python
for i in title:
    print(i.text)
```

```python
# 가수 백명 수집
singer = soup.select("div.ellipsis.rank02>a")
for i in singer:
    print(i.text)
```

```python
# 항상 데이터의 개수가 정확한지 검증작업이 필수!
# 개수가 107개인 이유는 a태그를 각각 하나씩 들고와서
# 가수가 여러명인경우 -> 해결책(가수전체를 포함한 태그를 수집)
print(len(singer))
```

```python
# 개수는 100개가 정확한데 내용이 중복
# 밑에 span태그가 보이진 않지만 태그가 존재! display: none
singer = soup.select("div.ellipsis.rank02")
len(singer)
```

```python
for i in singer:
    print(i.text)
```

```python
# span만 통으로 가지고 와야한다.
singer = soup.select("span.checkEllipsis")
len(singer)
```

```python
for i in singer:
    print(i.text)
```
