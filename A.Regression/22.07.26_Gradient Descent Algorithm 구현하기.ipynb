```python
import numpy as np
```

# 사용할 데이터셋 생성

## 문제 생성(공부한 시간)

```python
x = np.random.uniform(low = 0.0, high = 1.0, size = 100)
print(x.shape)
print(x)
```

## 정답생성(성적)

```python
y = 10 * x + 5
print(y.shape)
print(y)
```

# 경사하강법을 이용한 학습

```python
num_epoch = 500 # 경사하강법을 적용할 최대 반복 횟수
learning_rate = 1.0 # 학습률
# 초기 가중치와 절편을 셋팅
w = np.random.uniform(low = 0.0, high = 1.0)
b = np.random.uniform(low = 0.0, high = 1.0)

print('초기 가중치 : ', w)
print('초기 절편 : ',b)
```

```python
for epoch in range(num_epoch):
        pre_y = w * x + b
        mse = ((y - pre_y)**2).mean()
        print("MSE : ",mse)
        
        if mse< 0.001 :
            break
            
        # 경사하강법을이용해 w,b 업데이트
        w = w - learning_rate * ((pre_y - y)*x).mean()
        b = b -learning_rate * ((pre_y -y)*1).mean()
        
        print('epoch : ', (epoch + 1), 'w : ',w ,'b :',b)
    
```

# 입력특성(데이터 컬럼) 2개짜리 만들어보기

```python
x1 = np.random.uniform(low = 0.0, high = 1.0, size = 100)
x2 = np.random.uniform(low = 0.5, high = 1.5, size = 100)
```

```python
y = 13 * x1 + 5 * x2 -2
```

```python
num_epoch = 500 # 경사하강법을 적용할 최대 반복 횟수
learning_rate = 0.6 # 학습률
# 초기 가중치와 절편을 셋팅
w1 = np.random.uniform(low = 0.0, high = 1.0)
w2 = np.random.uniform(low = 0.0, high = 1.0)
b = np.random.uniform(low = 0.0, high = 1.0)

print('초기 가중치 w1 : ', w1)
print('초기 가중치 w2 : ', w2)
print('초기 절편 : ',b)
```

```python
for epoch in range(num_epoch):
        pre_y = (w1 * x1) + (w2 * x2 )+ b
        mse = ((y - pre_y)**2).mean()
        print("MSE : ",mse)
        
        if mse< 0.001 :
            break
            
        # 경사하강법을이용해 w,b 업데이트
        w1 = w1 - learning_rate * ((pre_y-y)*x1).mean()
        w2 = w2 - learning_rate * ((pre_y-y)*x2).mean()
        b = b -learning_rate * ((pre_y -y)*1).mean()
        
        print('epoch : ', (epoch + 1), 'w1 : ',w1 ,'w2 : ',w2,'b :',b)
```
