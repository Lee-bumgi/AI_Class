- 결측치 채우는 방법

## Age 결측치 채우기
- 결측치가 있는 데이터의 다른 컬럼의 값과 같은 데이터의 평균값을 사용해서 결측치를 채움
- 그럼 어떤 컬럼을 참조할까요?
    - 결측치가 있는 컬럼(Age)와 상관관계가 높은 컬럼 선택(범주형)
- 피벗 테이블을 활용
- apply()을 이용하여 전체 데이터의 결측치를 채움
    - DF의 data를 분리해서 원하는 처리를 수행한 후 다시 병합하는 함수. 재구조화 함수(reconstruct function)

```python
train.corr()
```

```python
# 피벗 테이블 생성
# values : 결측치를 채울 컬럼
# index : 참고할 컬럼 목록 리스트
# aggfunc : 사용할 수학 도구(평균, 중간값)
pt1 = train.pivot_table(values = "Age", index = ["Pclass","Sex"],aggfunc = "mean")
pt1
```

```python
# 피벗 테이블에 접근하는 방법
pt1.loc[1,"male"]
```

### 결측치 채우는 함수

```python
import numpy as np
# Age 컬럼의 결측치를 채우는 함수
def fill_age(row):
    # 한 줄 데이터에서 Age 컬럼이 결측치라면
    if np.isnan(row["Age"]):
        # 피벗 테이블을 참조 (같은 Pclass와 Sex인 값을 반환)
        return pt1.loc[row["Pclass"], row ["Sex"]]
    else:
        return row["Age"]
```

```python
# axis = 1 모든 컬럼들을 넘긴다.
# astype("int64") : 나이는 실수가 없으니 정수로 변환
train["Age"]= train.apply(fill_age, axis = 1).astype("int64")
```

```python
train
```

```python
test["Age"]= test.apply(fill_age, axis = 1).astype("int64")
test
```

```python
test.info()
```

## train 데이터의 Embarked결측치 채우기

```python
# 가장 많은 데이터수를 갖는 클래스(S)로 결측치를 할당
train["Embarked"].value_counts()
```

```python
# Embarked 컬럼의 결측치 값을 S로 채워라
train["Embarked"]= train["Embarked"].fillna("S")
train.info()
```

```python
test.info()
```

```python
## test 데이터의 Fare 결측치 채우기
```

```python
test.corr()
```

```python
pt2 = test.pivot_table(values = "Fare", index = ["Pclass","Embarked"],aggfunc = "mean")
pt2
```

```python
def fill_Fare(row):
    # 한 줄 데이터에서 Age 컬럼이 결측치라면
    if np.isnan(row["Fare"]):
        # 피벗 테이블을 참조 (같은 Pclass와 Sex인 값을 반환) - 범주형 데이터만 사용해야한다. Age같은 경우는 없는 값이 존재할수있따.
        return pt2.loc[row["Pclass"], row ["Sex"]]
    else:
        return row["Fare"]
test["Fare"]= test.apply(fill_Fare, axis = 1)
test
```

```python
test.info()
```

```python
test[test["Fare"].isnull()]
```

```python
test["Fare"] = test["Fare"].fillna(13.913030)
test["Fare"] = test["Fare"].fillna(float(pt2.loc[3,"S"]))
```

```python
# 일부 수치가 적은 호칭을 other로 통일하자.
# Mr, Mrs, Miss, Master, Rev, Other
```
