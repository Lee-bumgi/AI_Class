## 문제정의
- 보스턴 집값 데이터에서 어떤 특성이 집값에 얼마만큼 영향을 주는지 분석
- KNN을 이용한 회귀분석
- 다중 공선성 문제에 대해 이해

## 데이터 수집

```python
# sklearn 에서는 교육용 데이터로 보스턴 집값 데이터를 제공한다. (dictionary 형태)
from sklearn.datasets import load_boston

boston = load_boston()

boston.keys()
# data : 특성 데이터
# target : 라벨 데이터
# feature_names : 컬럼명
# DESCR : 컬럼에 대한 설명
```

## 데이터 시각화 및 분석
- 딕셔너리를 데이터프레임으로 변환
- 특성과 라벨 간의 상관관계 분석
- 분석 결과를 히트맵으로 시각화

```python
import pandas as pd

df_feature = pd.DataFrame(boston["data"],columns = boston["feature_names"])

df_feature.head()
```

```python
df_feature.info()
```

```python
df_feature.describe()
```

```python
df_label = pd.DataFrame(boston['target'],columns = ['price'])
df_label.head()
```

```python
# 특성 데이터와 라벨데이터를 하나의 데이터프레임으로 병합 -> concat()
df_boston = pd.concat([df_feature,df_label],axis=1)

df_boston
```

```python
# 특성과 라벨의 상관관계 분석
df_boston.corr()

# 양의 정수 : 집 값을 
```
